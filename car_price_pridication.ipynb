### Evaluating Regression Model Performance

For regression tasks, we use different metrics than for classification. Here are some common ones:

*   **Mean Absolute Error (MAE):** The average of the absolute differences between predictions and actual values. It gives an idea of the magnitude of errors. A lower MAE indicates a better model.

*   **Mean Squared Error (MSE):** The average of the squared differences between predictions and actual values. It penalizes larger errors more heavily than MAE. A lower MSE indicates a better model.

*   **Root Mean Squared Error (RMSE):** The square root of the MSE. It has the same units as the target variable, making it easier to interpret than MSE. A lower RMSE indicates a better model.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test_sc, y_pridict)
print(f'Mean Absolute Error (MAE): {mae:.4f}')

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test_sc, y_pridict)
print(f'Mean Squared Error (MSE): {mse:.4f}')

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
```

### Visualizing Model Performance: Actual vs. Predicted and Residuals

```python
# Inverse transform the scaled predictions and actual test values
y_pred_original = sc_y.inverse_transform(y_pridict)
y_test_original = sc_y.inverse_transform(y_test_sc)

# Create a DataFrame for easier plotting
results_df = pd.DataFrame({
    'Actual': y_test_original.flatten(),
    'Predicted': y_pred_original.flatten()
})

# Plot Actual vs. Predicted Values
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Actual', y='Predicted', data=results_df, alpha=0.6)
plt.plot([results_df['Actual'].min(), results_df['Actual'].max()],
         [results_df['Actual'].min(), results_df['Actual'].max()],
         color='red', linestyle='--', linewidth=2) # Ideal 45-degree line
plt.title('Actual vs. Predicted Selling Price (Original Scale)')
plt.xlabel('Actual Selling Price')
plt.ylabel('Predicted Selling Price')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calculate residuals
results_df['Residuals'] = results_df['Actual'] - results_df['Predicted']

# Plot Residuals vs. Predicted Values
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Predicted', y='Residuals', data=results_df, alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.title('Residuals vs. Predicted Selling Price (Original Scale)')
plt.xlabel('Predicted Selling Price')
plt.ylabel('Residuals (Actual - Predicted)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### How to interpret these plots:

*   **Actual vs. Predicted Plot:** If your model was perfect, all points would lie exactly on the red dashed line (where Actual = Predicted). The closer the points are to this line, the better your model's predictions.

*   **Residual Plot:** For a good model, the residuals should be randomly scattered around the zero line (red dashed line) with no discernible pattern. If you see patterns (e.g., a cone shape, a curve), it might indicate that the model is missing some important relationships or that there's heteroscedasticity (non-constant variance of errors).

### Setting up Encoders and Preprocessing for New Input

To allow input of original values, we need to re-create the `LabelEncoder` objects and the `StandardScaler` objects and apply the exact same preprocessing steps as were done for the training data.

First, we'll reload the original dataset and apply the `km_driven` and `selling_price` outlier removal to match the `df` used for training. This ensures our encoders are fitted on the correct data distribution.

```python
# Reload original data to fit encoders correctly
df_original = pd.read_csv("/content/CAR DETAILS FROM CAR DEKHO.csv")

# Apply the same km_driven outlier removal
q1_km = df_original.km_driven.quantile(0.25)
q3_km = df_original.km_driven.quantile(0.75)
iqr_km = q3_km - q1_km
lower_km = q1_km - 1.5 * iqr_km
upper_km = q3_km + 1.5 * iqr_km
df_processed_for_encoders = df_original[(df_original.km_driven >= lower_km) & (df_original.km_driven <= upper_km)].copy()

# Apply the same selling_price outlier removal
q1_price = df_processed_for_encoders.selling_price.quantile(0.25)
q3_price = df_processed_for_encoders.selling_price.quantile(0.75)
iqr_price = q3_price - q1_price
lower_price = q1_price - 1.5 * iqr_price
upper_price = q3_price + 1.5 * iqr_price
df_processed_for_encoders = df_processed_for_encoders[(df_processed_for_encoders.selling_price >= lower_price) & (df_processed_for_encoders.selling_price <= upper_price)].copy()

# Instantiate and fit separate LabelEncoders for each categorical column and 'year'
le_name = LabelEncoder()
le_year = LabelEncoder()
le_fuel = LabelEncoder()
le_seller_type = LabelEncoder()
le_transmission = LabelEncoder()
le_owner = LabelEncoder()

le_name.fit(df_processed_for_encoders['name'])
le_year.fit(df_processed_for_encoders['year'])
le_fuel.fit(df_processed_for_encoders['fuel'])
le_seller_type.fit(df_processed_for_encoders['seller_type'])
le_transmission.fit(df_processed_for_encoders['transmission'])
le_owner.fit(df_processed_for_encoders['owner'])

print("Encoders have been fitted successfully.")
```

### Updated Prediction Function (Takes Original Values)

Now, here's the modified `predict_new_car_price` function. You can directly pass the raw values for each feature, and it will handle the encoding, year transformation (log then sqrt), and scaling internally.

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pandas as pd
import numpy as np

def predict_new_car_price_original_input(name, year, km_driven, fuel, seller_type, transmission, owner):
    """
    Predicts the selling price of a car given its original, untransformed features.

    Args:
        name (str): Original string value for 'name'.
        year (int): Original integer value for 'year'.
        km_driven (int): Original integer value for 'km_driven'.
        fuel (str): Original string value for 'fuel'.
        seller_type (str): Original string value for 'seller_type'.
        transmission (str): Original string value for 'transmission'.
        owner (str): Original string value for 'owner'.

    Returns:
        float: Predicted selling price in its original scale.
    """
    # 1. Label Encode the categorical and year features
    encoded_name = le_name.transform([name])[0]
    encoded_year = le_year.transform([year])[0]
    encoded_fuel = le_fuel.transform([fuel])[0]
    encoded_seller_type = le_seller_type.transform([seller_type])[0]
    encoded_transmission = le_transmission.transform([transmission])[0]
    encoded_owner = le_owner.transform([owner])[0]

    # 2. Apply log and sqrt transformations to the encoded year
    # Handle potential log(0) if encoded_year is 0
    transformed_year = np.log(encoded_year) if encoded_year > 0 else 0 # Replace with 0 if log(0)
    transformed_year = transformed_year**(1/2)

    # Create a DataFrame for the single new input instance
    new_data = pd.DataFrame([{
        'name': encoded_name,
        'year': transformed_year,
        'km_driven': km_driven,
        'fuel': encoded_fuel,
        'seller_type': encoded_seller_type,
        'transmission': encoded_transmission,
        'owner': encoded_owner
    }])

    # 3. Scale the new input features using the fitted sc_X scaler
    new_data_scaled = sc_X.transform(new_data)

    # 4. Handle potential NaNs in scaled input, as done during training
    new_data_scaled_clean = np.nan_to_num(new_data_scaled, nan=0.0)

    # 5. Make prediction using the trained linear model
    scaled_prediction = linear.predict(new_data_scaled_clean)

    # 6. Inverse transform the prediction to get the price in original scale
    original_price_prediction = sc_y.inverse_transform(scaled_prediction.reshape(-1, 1))

    return original_price_prediction[0][0]

# --- Example Usage (Input original, human-readable values) ---
# You can find valid values by checking your original 'df_original' DataFrame or the value_counts() from earlier cells.

predicted_price_new = predict_new_car_price_original_input(
    name='Maruti 800 AC', # Example: An actual car name from your dataset
    year=2007,           # Example: An actual year from your dataset
    km_driven=70000,     # Example: Kilometer driven
    fuel='Petrol',       # Example: 'Petrol', 'Diesel', 'CNG', 'LPG', 'Electric'
    seller_type='Individual', # Example: 'Individual', 'Dealer', 'Trustmark Dealer'
    transmission='Manual', # Example: 'Manual', 'Automatic'
    owner='First Owner'  # Example: 'First Owner', 'Second Owner', 'Third Owner', 'Fourth & Above Owner'
)

print(f"Predicted Selling Price (Original Input): ₹{predicted_price_new:,.2f}")
```
